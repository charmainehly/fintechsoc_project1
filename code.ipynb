# 1. Linear Regression

More for understanding how to implement linear regression and bringing in other variables that I think could be important. Not a good prediction model to be used because

```python
from sklearn.linear_model import LinearRegression
from datetime import datetime

model = LinearRegression()

historical_data['Date'] = pd.to_datetime(historical_data.Date, format='%Y-%m-%d')
```

```python
data_model_1 = historical_data.iloc[:, [0,4]].copy()
data_model_1.index = historical_data['Date']

data_model_1.head()
```

```python
# loop through to generate variables values for 
# ‘Year’, ‘Month’, ‘Week’, ‘Day’, ‘Dayofweek’, ‘Dayofyear’, 
# ‘Is_month_end’, ‘Is_month_start’, ‘Is_year_end’, and  ‘Is_year_start’.
# Hypothesis on quarter on the year affecting performance, as well as start/end of week

data_model_1['year'] = 0
data_model_1['month'] = 0
data_model_1['day'] = 0
data_model_1['dayofweek'] = 0
data_model_1['mon_fri'] = 0
data_model_1['dayofyear'] = 0
data_model_1['quarter'] = 0
data_model_1['Is_year_end'] = 0
data_model_1['Is_year_start'] = 0

data_model_1['year'] = pd.DatetimeIndex(data_model_1['Date']).year
data_model_1['month'] = pd.DatetimeIndex(data_model_1['Date']).month
data_model_1['day'] = pd.DatetimeIndex(data_model_1['Date']).day
data_model_1['dayofweek'] = data_model_1['Date'].dt.dayofweek
data_model_1['dayofyear'] = data_model_1['Date'].dt.strftime('%j')


for i in range(0,len(data_model_1)):
#mon_fri
    if (data_model_1['dayofweek'][i] == 0 or data_model_1['dayofweek'][i] == 4):
        data_model_1.iloc[i,6] = 1
    else:
        data_model_1.iloc[i,6] = 0

    if (data_model_1['month'][i] == 1 or data_model_1['month'][i] == 2 or data_model_1['month'][i] == 3):
        data_model_1.iloc[i,8] = 1
    elif (data_model_1['month'][i] == 4 or data_model_1['month'][i] == 5 or data_model_1['month'][i] == 6):
        data_model_1.iloc[i,8] = 2
    elif (data_model_1['month'][i] == 7 or data_model_1['month'][i] == 8 or data_model_1['month'][i] == 9):
        data_model_1.iloc[i,8] = 3
    else:
        data_model_1.iloc[i,8] = 4
    
    if (data_model_1['month'][i] == 1 and data_model_1['day'][i] == 1):
        data_model_1.iloc[i,10] = 1
    else:
        data_model_1.iloc[i,10] = 0

    if (data_model_1['month'][i] == 12 and data_model_1['day'][i] == 12):
        data_model_1.iloc[i,9] = 1
    else:
        data_model_1.iloc[i,9] = 0

#data_model_1['Date'] = data_model_1['Date'].astype(str)

data_model_1.head()
```

```python
#training data
data_model_1 = data_model_1.drop(['Date'], axis=1)
x_train = np.array(data_model_1.drop('Close', axis=1))
y_train = np.array(data_model_1['Close'])

#implement linear regression
model.fit(x_train,y_train)
```

# 2. Prophet

```python
from prophet import Prophet

# Reading up different models from https://www.analyticsvidhya.com/blog/2018/10/predicting-stock-price-machine-learningnd-deep-learning-techniques-python/
# On different models utilized for stock price prediction and pros/cons
```

```python
#creating dataframe
historical_data['Date'] = pd.to_datetime(historical_data.Date, format='%Y-%m-%d')
data_model_2 = historical_data.iloc[:, [0,4]].copy()
data_model_2.index = historical_data['Date']
data_model_2.head()
```

```python
#preparing data
data_model_2.rename(columns={'Close': 'y', 'Date': 'ds'}, inplace=True)

#fit the model
model = Prophet(daily_seasonality=True)
model.fit(data_model_2)
```

```python
sample = pd.read_csv("sample_submission.csv")
prediction_model_2 = sample.iloc[:, [0]].copy()
prediction_model_2['Date'] = pd.to_datetime(prediction_model_2.Date, format='%Y-%m-%d')
prediction_model_2.index = prediction_model_2['Date']

prediction_model_2.head()
```

```python
#predictions
predicted = model.make_future_dataframe(periods=len(prediction_model_2))
forecast = model.predict(predicted)

forecast_predicted = forecast['yhat'][1628:]
prediction_model_2['Predicted'] = forecast_predicted.values
prediction_model_2.head()
```

```python
prediction_model_2.to_csv('submission_attempt.csv',index=False)
```
